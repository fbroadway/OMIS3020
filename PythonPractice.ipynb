program that will splice a string with the word "lucky" in it, and just print "lucky" no matter where it is. Search string len() times, checking if the first 5 characters are "lucky" and moving to the next one if not.

```python
s="Today is your lucky day!"
k = "lucky"

for i in range(len(s)):
  printString = ""
  for j in range(5):
    if s[i+j] == k[j]:
      printString = printString + s[i+j]
      j += 1
    else:
      break
  if printString == k:
    print(printString)
    print("The word lucky was found at index #" + str(i))
    break
  else:
    i += 1
if i == range(len(s)):
  print("This string does not contain the word 'Lucky'")

```
