Data Preprocessing Steps: Data Exploration (Desc. Stats, DataViz) -> Data Cleaning (Remove Dupes, Handle Missing Vals and Outliers) -> Data Transformation (Scaling, Encoding) -> Data Integration (Joining, Merging) -> Data Reduction (Dimensionality Reduction)

stats babey

measures of central tendency

-mean good for quantitative data

-median good for same data types as mean, but better for summarizing data when outliers exist

-mode good on numerical or categorical


skewness

-right skew (higher on left) - mean > median > mode - positive outliers

-left skew (long tail on left) - mean < median < mode - negative outliers

measures of spread/dispersion

-range (max - min) - does not accoutn for outliers

-variance - how far is sum of squared distances from each point to the mean

-standard deviation - more informative due to being on similar scale to the data you're working with

measures of [how similar things are to each other]

-covariance - how do changes in one variable affect the other (only sign is meaningful)

-correlation - normalizes covariance to glean effective information

```python
over_60 = dfTitanic['age'] > 60
cabin_exists = dfTitanic['cabin'].notnull()
```

```python
#combining booleans - AND and OR not useable with dataframes, & replaces AND, +/| replaces OR
over_60_and_cabin_exists = over_60 & cabin_exists
over_60_and_cabin_exists
```

```python
over_60_or_cabin_exists = over_60 + cabin_exists
over_60_or_cabin_exists
```
