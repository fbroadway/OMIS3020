```python
#%%
import os
import subprocess
from google.colab import drive
import getpass

# Mount Google Drive
drive.mount("/content/drive")

# Repository settings (EDIT THESE)
repository = "OMIS3020"        # enter your repository name
username = "fbroadway"       # enter your GitHub username
email = "francesco.broadway@gmail.com"  # enter your GitHub email (or noreply email)
name = "Francesco Broadway"     # enter your full name (for commit history)

# Enter your GitHub token (hidden input)
token = getpass.getpass("Paste your GitHub token (hidden): ").strip()

# Where you want the repo to live in Drive
folder_path = f"/content/drive/MyDrive/{repository}"

# Make sure we're in MyDrive and it exists
mydrive_path = "/content/drive/MyDrive"
if not os.path.exists(mydrive_path):
    print(f"Error: {mydrive_path} doesn't exist. Is Drive mounted correctly?")
else:
    os.chdir(mydrive_path)
    print(f"Current directory: {os.getcwd()}")

    # Clone only if needed
    if not os.path.exists(folder_path):
        clone_url = f"https://x-access-token:{token}@github.com/{username}/{repository}.git"
        print(f"Cloning into {folder_path}...")
        result = os.system(f'git clone "{clone_url}"')

        if result == 0:
            print(" Clone successful!")
        else:
            print(" Clone failed")
    else:
        print(f"Repo already exists: {folder_path}")

    # Move into the repo
    if os.path.exists(folder_path):
        os.chdir(folder_path)
        print(f"Now in: {os.getcwd()}")

        # Set Git identity
        subprocess.run(["git", "config", "--global", "user.email", email], check=True)
        subprocess.run(["git", "config", "--global", "user.name", name], check=True)
        print(" Git configured")
    else:
        print(f"Error: Could not find {folder_path}")
```
